```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Properties.csv')

def clean_data(df):
    sqm_to_sqft = 10.7639

    clean_area= df.area.str.replace(',','').str.replace('sq. ft', '').astype(float) / sqm_to_sqft
    df['sqft_in_sqm'] = df.sqft / sqm_to_sqft
    best_parsed = df[['sqft_in_sqm','sqm']].mean(axis=1)

    df['link'] = df.link.str.replace('//','/')

    df['best_sqm'] = np.where(pd.isnull(clean_area), best_parsed, clean_area)

    df = df.drop(columns=['area', 'Unnamed: 0'])

    df.to_csv('processed_data.csv')

    return (df.sort_values('best_sqm',ascending=False))

clean_data(df).head(10)
```

```python
import os
import json

files = os.listdir('json_data')

list_data = []

for f in files:
    with open('json_data/'+f) as js:
        data = json.load(js)
        list_data.append(data)

pd.DataFrame(list_data)
```

```python
df['latitude'] = df['gmap'].str.extract("'latitude': (\d\d\.\d{2,6})").astype(float)
df['longitude'] = df['gmap'].str.extract("'longitude': (-{0,1}\d.\d{2,6})").astype(float)

clean_data(df).head(10)
```

```python
import geopandas

gdf = geopandas.GeoDataFrame(df, geometry= geopandas.points_from_xy(df.longitude, df.latitude))
```

```python
gdf.set_crs(crs='EPSG:4326', inplace=True)
gdf[(df.best_sqm > 72) & (df.best_sqm < 150 )].explore("best_sqm", tiles = "CartoDB dark_matter", popup = True)
```
